```python
def fill_with_previous(series):
    last_valid = None
    for idx, value in series.items():
        if value == 0 and last_valid is not None:
            series[idx] = last_valid
        elif value != 0:
            last_valid = value
    return series
vwap = fill_with_previous()
```
